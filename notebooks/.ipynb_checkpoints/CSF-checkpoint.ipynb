```python
use_cuda = True
from scvi.harmonization.utils_chenling import get_matrix_from_dir
import numpy as np
from scvi.dataset.dataset import GeneExpressionDataset
from scvi.harmonization.utils_chenling import eval_latent
from scvi.models.vae import VAE
from scvi.inference import UnsupervisedTrainer
from copy import deepcopy



class MSDataset(GeneExpressionDataset):
    def __init__(self, patient='MS19270',celltype='PBMCs'):
        self.patient = patient
        self.celltype = celltype
        count, labels, cell_type, gene_names,cellid = self.preprocess()
        self.cellid = cellid
        super(MSDataset, self).__init__(
            *GeneExpressionDataset.get_attributes_from_matrix(
                count, labels=labels),
            gene_names=np.char.upper(gene_names), cell_types=cell_type)
    def preprocess(self):
        print("Preprocessing data")
        count, geneid, cellid = get_matrix_from_dir('Gerd/{0}/{1}/GRCh38'.format(self.patient,self.celltype), storage='/data/')
        geneid = geneid[:, 1]
        count = count.T.tocsr()
        print("Finish preprocessing data")
        labels = np.repeat(0,len(cellid))
        cell_type = ['unlabelled']
        return count,labels,cell_type,geneid,cellid
```

```python

class CSFDataset(GeneExpressionDataset):
    def __init__(self):
        count, labels, cell_type, gene_names,cellid = self.preprocess()
        super(MSDataset, self).__init__(
            *GeneExpressionDataset.get_attributes_from_matrix(
                count.tocsr(), labels=labels),
            gene_names=np.char.upper(gene_names), cell_types=cell_type,cell_id = cellid)
    def preprocess(self):
        print("Preprocessing data")
        count, geneid, cellid = get_matrix_from_dir('CSF',storage='/data/')
        print("Finish preprocessing data")
        labels = np.repeat(0,len(cellid))
        cell_type = ['unlabelled']
        return count,labels,cell_type,geneid,cellid



CSF = MSDataset()
celltypes = pd.read_csv('/data/CSF/celltypes.txt')
celltype = np.asarray(celltypes['celltypes'])
donor = np.asarray(celltypes['donor'])

CSF.cell_types,CSF.labels = np.unique(celltype,return_inverse=True)
CSF.batch_names,CSF.batch_indices = np.unique(donor,return_inverse=True)
CSF.batch_indices = CSF.batch_indices.reshape(len(CSF.batch_indices),1)
```

```python
def PlotMarkers(markers, coeff, gene_dataset, nrows,ncol,celltype,n_labelled=50,plot_exprs=False):
    marker_exprs =[np.asarray(gene_dataset.X[:,gene_dataset.gene_names==x].todense()).ravel() for x in markers]
    score = np.mean(np.asarray([x/np.max(x)*coeff[i] for i,x in enumerate(marker_exprs)]),axis=0)
    marker_exprs = marker_exprs+[score]
    plt.figure(figsize=(3*ncol, 3*nrows), facecolor='w', edgecolor='k')
    if plot_exprs ==True:
        for i, x in enumerate(markers+['score']):
            plt.subplot(nrows, ncol, i+1,facecolor='lightgray')
            plt.scatter(latent_u[:, 0], latent_u[:, 1],c=marker_exprs[i], cmap = sns.cubehelix_palette(as_cmap=True, dark=0),edgecolors='none',s=0.9)
            plt.axis("off")
            plt.tight_layout()
            if i == len(markers):
                plt.title('%s score' % celltype)
            elif coeff[i]<0:
                plt.title('%s negative marker:' % celltype +x)
            else:
                plt.title('%s positive marker:' % celltype +x)
            plt.tick_params(which='both',bottom=False,top=False,left=False, labelleft=False,labelbottom=False)
        plt.colorbar()
        plt.show()

        idx = np.argsort(score*-1)[:n_labelled]
        fig, ax = plt.subplots(figsize=(6, 6))
        ax.scatter(latent_u[:, 0], latent_u[:, 1],edgecolors='none',c='lightgray',alpha=0.6)
        ax.scatter(latent_u[idx, 0], latent_u[idx, 1],edgecolors='none',c='red',alpha=0.6)
        plt.title('%s %s are labeld' % (len(idx),celltype))
        plt.show()
    return(idx)
```
