```python
use_cuda = True
from scvi.harmonization.utils_chenling import get_matrix_from_dir
from scvi.dataset.pbmc import PbmcDataset
from scvi.harmonization.utils_chenling import assign_label
import numpy as np
from scipy.stats import entropy
from sklearn.neighbors import NearestNeighbors

from scvi.dataset.dataset import GeneExpressionDataset
from scvi.harmonization.utils_chenling import run_model
from copy import deepcopy

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
import matplotlib.pyplot as plt
import seaborn as sns
from umap import UMAP
from scvi.metrics.clustering import clustering_scores,select_indices_evenly

from scvi.models.vae import VAE
from scvi.models.scanvi import SCANVI
from scvi.inference import UnsupervisedTrainer, SemiSupervisedTrainer
import torch

from scvi.harmonization.utils_chenling import SubsetGenes
```

```python
dataset1 = PbmcDataset(filter_out_de_genes=False)
dataset1.update_cells(dataset1.batch_indices.ravel()==0)
dataset1.subsample_genes(dataset1.nb_genes)

count, geneid, cellid = get_matrix_from_dir('cite')
count = count.T.tocsr()
seurat = np.genfromtxt('../cite/cite.seurat.labels', dtype='str', delimiter=',')
cellid = np.asarray([x.split('-')[0] for x in cellid])
labels_map = [0, 0, 1, 2, 3, 4, 5, 6]
labels = seurat[1:, 4]
cell_type = ['CD4 T cells', 'NK cells', 'CD14+ Monocytes', 'B cells','CD8 T cells', 'FCGR3A+ Monocytes', 'Other']
dataset2 = assign_label(cellid, geneid, labels_map, count, cell_type, seurat)
```

### One population only contains B cells and the other contains all the other cell types

```python
rmCellTypes='B cells'
pbmc = deepcopy(dataset1)
newCellType = [k for i, k in enumerate(dataset1.cell_types) if k not in [rmCellTypes]]
pbmc.filter_cell_types(newCellType)
pbmc2 = deepcopy(dataset2)
pbmc2.filter_cell_types([rmCellTypes])
```

```python
pbmc.cell_types
```

```python
pbmc2.cell_types
```

```python
gene_dataset = GeneExpressionDataset.concat_datasets(pbmc, pbmc2)
gene_dataset.subsample_genes(1000)
```

```python
rerun=False
vae = VAE(gene_dataset.nb_genes, n_batch=gene_dataset.n_batches,n_hidden=128, n_latent=10, n_layers=2, dispersion='gene')
trainer = UnsupervisedTrainer(vae, gene_dataset, train_size=1.0)
if os.path.isfile('../NoOverlap/vae.scvigenes.Bcells.pkl')  and rerun==False:
    trainer.model.load_state_dict(torch.load('../NoOverlap/vae.scvigenes.Bcells.pkl'))
    trainer.model.eval()
else:
    trainer.train(n_epochs=150)
    torch.save(trainer.model.state_dict(),'../NoOverlap/vae.scvigenes.Bcells.pkl')

full = trainer.create_posterior(trainer.model, gene_dataset, indices=np.arange(len(gene_dataset)))
```

```python
labels = deepcopy(gene_dataset.labels.ravel())
true_labels = deepcopy(labels)
true_celltypes = deepcopy(gene_dataset.cell_types)
rmidx = np.where(gene_dataset.cell_types==rmCellTypes)[0]
tempidx = np.where(gene_dataset.cell_types=='NK cells')[0]
labels[labels==rmidx] = tempidx
```

```python
true_celltypes
```

```python
newtype,newlabel = np.unique(labels,return_inverse=True)
gene_dataset.cell_types = gene_dataset.cell_types[newtype]
labels = newlabel.reshape(len(newlabel),1)
gene_dataset.labels = labels
```

```python
scanvi = SCANVI(gene_dataset.nb_genes, gene_dataset.n_batches, (pbmc.n_labels+1), n_layers=2,
               classifier_parameters = {'dropout_rate':0.5, 'n_hidden':256, 'n_layers':2})
scanvi.load_state_dict(trainer.model.state_dict(), strict=False)
trainer_scanvi = SemiSupervisedTrainer(scanvi, gene_dataset, classification_ratio=50,
                                       n_epochs_classifier=1, lr_classification=5 * 1e-3)
```

```python
trainer_scanvi.labelled_set = trainer_scanvi.create_posterior(indices = gene_dataset.batch_indices.ravel()==0)
trainer_scanvi.unlabelled_set = trainer_scanvi.create_posterior(indices = gene_dataset.batch_indices.ravel()==1)
```

```python
if os.path.isfile('../NoOverlap/scanvi.scvigenes.Bcells.pkl') and rerun==False:
    trainer_scanvi.model.load_state_dict(torch.load('../NoOverlap/scanvi.scvigenes.Bcells.pkl'))
    trainer_scanvi.model.eval()
else:
    trainer_scanvi.train(n_epochs=10)
    torch.save(trainer_scanvi.model.state_dict(),'../NoOverlap/scanvi.scvigenes.Bcells.pkl')
```

```python
scanvi_false = SCANVI(gene_dataset.nb_genes,gene_dataset.n_batches,(pbmc.n_labels), n_layers=2,
                     classifier_parameters = {'dropout_rate':0.5, 'n_hidden':256, 'n_layers':2})
scanvi_false.load_state_dict(trainer.model.state_dict(), strict=False)
```

```python
trainer_scanvi_false = SemiSupervisedTrainer(scanvi_false, gene_dataset, classification_ratio=50,
                       n_epochs_classifier=1, lr_classification=5 * 1e-3)

trainer_scanvi_false.labelled_set = trainer_scanvi_false.create_posterior(indices = gene_dataset.batch_indices.ravel()==0)
trainer_scanvi_false.unlabelled_set = trainer_scanvi_false.create_posterior(indices = gene_dataset.batch_indices.ravel()==1)
```

```python
if os.path.isfile('../NoOverlap/scanvi_f.scvigenes.Bcells.pkl') and rerun==False:
    trainer_scanvi_false.model.load_state_dict(torch.load('../NoOverlap/scanvi_f.scvigenes.Bcells.pkl'))
    trainer_scanvi_false.model.eval()
else:
    trainer_scanvi_false.train(n_epochs=10)
    torch.save(trainer_scanvi_false.model.state_dict(),'../NoOverlap/scanvi_f.scvigenes.Bcells.pkl')
```

### Check that the VAE latent space makes sense

```python
latent, batch_indices, _ = full.sequential().get_latent()
keys = true_celltypes
key_order = np.argsort(keys)
labels = true_labels

sample = select_indices_evenly(2000, labels)
colors = sns.color_palette('tab20')
latent_s = latent[sample, :]
label_s = true_labels[sample]
batch_s = batch_indices.ravel()[sample]

from umap import UMAP
latent_u = UMAP(spread=2).fit_transform(latent_s)

fig, ax = plt.subplots(figsize=(8, 8))
for i,k in enumerate(key_order):
    idx = label_s==k
    ax.scatter(latent_u[idx, 0], latent_u[idx, 1], c=colors[i%20], label=keys[k],
                   edgecolors='none')
plt.legend(bbox_to_anchor=(1.1, 0.5), borderaxespad=0, fontsize='x-large')
plt.axis("off")
fig.tight_layout()
plt.savefig("figures/SCANVI_stress_vaeUMAP_labels.pdf", transparency=True)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
for i,k in enumerate(['all others (PBMC8K)','B only (Cite-Seq) ']):
    idx = batch_s==i
    ax.scatter(latent_u[idx, 0], latent_u[idx, 1], c=colors[i%20], label=k,
                   edgecolors='none')
plt.legend(bbox_to_anchor=(1.1, 0.5), borderaxespad=0, fontsize='x-large')
plt.axis("off")
fig.tight_layout()
plt.savefig("figures/SCANVI_stress_vaeUMAP_batch.pdf", transparency=True)
```
