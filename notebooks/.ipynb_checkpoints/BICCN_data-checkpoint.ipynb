```python
dataset1 = Zeng10X(coarse=False)
```

```python
dict1 = dict(zip(dataset1.cell_types, dataset1.groups[dataset1.labels_groups]))
```

```python
dataset2 = ZengSS2(coarse=False)
```

```python
dict2 = dict(zip(dataset2.cell_types, dataset2.groups[dataset2.labels_groups]))
```

```python
dataset1.subsample_genes(dataset1.nb_genes)
dataset2.subsample_genes(dataset2.nb_genes)
```

```python
gene_dataset = GeneExpressionDataset.concat_datasets(dataset1, dataset2)
```

```python
from scvi.harmonization.utils_chenling import run_model
gene_dataset.subsample_genes(1000)

latent, batch_indices, labels, keys, stats = run_model('scanvi2', gene_dataset, dataset1, dataset2,
                                                       filename='Zeng', rep='10Xnuclei_SS2cell')
```

```python
colors = sns.color_palette('bright') +\
sns.color_palette('muted') + \
sns.color_palette('dark') + \
sns.color_palette('pastel') + \
sns.color_palette('colorblind')
```

```python
sample = select_indices_evenly(2000, labels)
latent_s = latent[sample, :]
label_s = labels[sample]
batch_s = batch_indices[sample]
if latent_s.shape[1] != 2:
    latent_s = UMAP(spread=2).fit_transform(latent_s)

key_order = np.argsort(keys)
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
for i,k in enumerate(key_order):
    ax.scatter(latent_s[label_s == k, 0], latent_s[label_s == k, 1], c=colors[i%30], label=keys[k],
               edgecolors='none')
    ax.legend(bbox_to_anchor=(1.1, 0.5), borderaxespad=0, fontsize='x-large')
ax.axis('off')
fig.show()
```

```python
group1 = dict(zip(dataset1.cell_types,dataset1.groups[dataset1.labels_groups]))
```

```python
group2 = dict(zip(dataset2.cell_types,dataset2.groups[dataset2.labels_groups]))
```

```python
group_dict = {**group1,**group2}
```

```python
groups = [group_dict[x] for x in gene_dataset.cell_types[gene_dataset.labels.ravel()]]
```

```python
groups,labels_g = np.unique(groups,return_inverse=True)
```

```python
label_gs = labels_g[sample]
print(keys[label_s][groups[label_gs]=='L5 IT TCAP'])
```

```python
label_gs = labels_g[sample]
groups_order = np.argsort(groups)

fig, ax = plt.subplots(figsize=(20,20))
for i,k in enumerate(groups_order):
    ax.scatter(latent_s[label_gs == k, 0], latent_s[label_gs == k, 1], c=colors[i%30], label=groups[k],
               edgecolors='none')
    ax.legend(bbox_to_anchor=(1.1, 0.5), borderaxespad=0, fontsize='x-large')
ax.axis('off')
fig.show()
```

```python
batch_id = ['10X','SS2']
batch_colors = ['blue','yellow']
fig, ax = plt.subplots(figsize=(20,20))
for i in range(2):
    ax.scatter(latent_s[batch_s == i, 0], latent_s[batch_s == i, 1], c=batch_colors[i], label=batch_id[i],
               edgecolors='none')
    ax.legend(bbox_to_anchor=(1.1, 0.5), borderaxespad=0, fontsize='x-large')
ax.axis('off')
fig.tight_layout()
fig.save('../batchlabels.pdf')
fig.show()
```
