```python
plotname = 'Tech3'
from scvi.harmonization.utils_chenling import trainVAE, SCANVI,SemiSupervisedTrainer
vae_posterior = trainVAE(all_dataset,filename=plotname,rep='0')
```

```python
from scvi.harmonization.utils_chenling import VAEstats
latent, batch_indices, labels, stats = VAEstats(vae_posterior)
```

```python
keys = all_dataset.cell_types
```

```python
colors = sns.color_palette('bright') +\
sns.color_palette('muted') + \
sns.color_palette('dark') + \
sns.color_palette('pastel') + \
sns.color_palette('colorblind')


sample = select_indices_evenly(2000, batch_indices)
latent_s = latent[sample, :]
label_s = labels[sample]
batch_s = batch_indices[sample]
if latent_s.shape[1] != 2:
    latent_s = UMAP().fit_transform(latent_s)

key_order = np.argsort(keys)
```

```python

fig, ax = plt.subplots(figsize=(10, 10))
for i,k in enumerate(key_order):
    ax.scatter(latent_s[label_s == k, 0], latent_s[label_s == k, 1], c=colors[i%30], label=keys[k],
               edgecolors='none')
    ax.legend(bbox_to_anchor=(1.1, 0.5), borderaxespad=0, fontsize='x-large')
ax.axis('off')
fig.show()
fig.savefig('../Tech3/celltype_label.pdf')
```

```python
batch=['10x','SS2']
fig, ax = plt.subplots(figsize=(10, 10))
for i,x in enumerate(batch):
    ax.scatter(latent_s[batch_s == i, 0], latent_s[batch_s == i, 1], c=colors[i], label=x,
               edgecolors='none')
    ax.legend(bbox_to_anchor=(1.1, 0.5), borderaxespad=0, fontsize='x-large')
ax.axis('off')
fig.show()
fig.savefig('../Tech3/batch_label.pdf')
```
